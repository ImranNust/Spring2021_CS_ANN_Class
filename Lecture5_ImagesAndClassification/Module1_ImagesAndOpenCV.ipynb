# What is OpenCV?
OpenCV is a Python open-source library, which is used for computer vision in Artificial intelligence, Machine Learning, face recognition, etc.

In OpenCV, the CV is an abbreviation form of a computer vision, which is defined as a field of study that helps computers to understand the content of the digital images such as photographs and videos.

The purpose of computer vision is to understand the content of the images. It extracts the description from the pictures, which may be an object, a text description, and three-dimension model, and so on. For example, cars can be facilitated with computer vision, which will be able to identify and different objects around the road, such as traffic lights, pedestrians, traffic signs, and so on, and acts accordingly.

We can retrieve a pixel value by its row and column coordinates. It returns an array of blue, green, red values of the BGR image. It returns the corresponding intensity for the grayscale image. First, we need to load the BGR image.

```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
 
I = np.array([[170,28,85,255,221,0],
              [68, 138, 17, 170, 119, 68],
             [221,0,238,136,0,255],
             [238, 17, 221, 68, 119, 225],
             [85, 170, 119, 221, 17, 136]],np.uint8)

# Finding out the size(shape) of the image
M, N = I.shape
print('There are',M,'rows, and there are',N,'columns')

# Accessing each element
for i in range(0,M):
    for j in range(0,N):
        print('The intensity value at location (', i, j, ') = ', I[i,j])
```

### Accessing 3x3 blocks of image

```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
 
I = np.array([[170,28,85,255,221,0],
              [68, 138, 17, 170, 119, 68],
             [221,0,238,136,0,255],
             [238, 17, 221, 68, 119, 225],
             [85, 170, 119, 221, 17, 136]],np.uint8)

# Finding out the size(shape) of the image
M, N = I.shape
print('There are',M,'rows, and there are',N,'columns')
k = 0
print("I = ", I)
# Accessing each element
for i in range(0,M-2):
    for j in range(0,N-2):
        print('Block'+str(k),'=', I[i:i+3,j:j+3])
        k+=1
```

### Padding 
***


```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
 
I = np.array([[170,283,85,255,221,0],
              [68, 138, 17, 170, 119, 68],
             [221,0,238,136,0,255],
             [238, 17, 221, 68, 119, 225],
             [85, 170, 119, 221, 17, 136]],np.uint8)

# Pad one each side of image I
I_pad = np.pad(I,(1,1),'constant',constant_values=1)
print(I_pad)
print("-------------------------------------------")
# Pad zero each side of image I
I_pad = np.pad(I,(1,1),'constant',constant_values=0)
print(I_pad)
print("-------------------------------------------")
# Repeat Edges on each side
I_pad = np.pad(I,(1,1),'edge')
print(I_pad)
print("-------------------------------------------")
# Pad 2 rows on each side using reflect method
I_pad = np.pad(I,(2,2),'reflect')
print(I_pad)
```

### Accessing different parameters of an image, such as
#### Shape (Size)
The number of rows and columns (and in case of color images, the number of channels as well) is called shape or size of an image.
####  Dimension of an image
For color images the dimension is 3, and for gray scale image, it is 2. 
####  Minimum Element
At the name suggest, the minimum value of an image.
####  Maximum Element
At the name suggest, the maximum value of an image.
####  Mean
The average of an image is called its mean.  
####  Median
The middle number in a sorted array of an image is called its median.
####  Mode
The most frequent value in an image is called its mode.
***

```python
# Accessing different parameters of a grayscale image
import numpy as np
import cv2 
from matplotlib import pyplot as plt

 
I = np.array([[170,28,85,255,221,0],
              [68, 138, 17, 170, 119, 68],
             [221,0,238,136,0,255],
             [238, 17, 221, 68, 119, 225],
             [85, 170, 119, 221, 17, 136]],np.uint8)

# The dimension of the image I
ndims = I.ndim
print("The dimension of the image 'I' is ", ndims)
print("-------------------------------------------")

# The shape of the image I
if ndims==2:
    print('It is a grayscale image')
    M, N = I.shape
    print('There are ',M, 'rows')
    print('There are ',N, 'columns')
elif ndims==3:
    print('It is a color image')
    M, N, nchannels = I.shape
    print('There are ',M, 'rows')
    print('There are ',N, 'columns')
    print('There are ',nchannels, 'channels')
else:
    print('Go! Correct yourself!')
print("-------------------------------------------")

# minimum value of the image I
I_min = np.min(I)
print("The minimum value of I is ", I_min)
print("-------------------------------------------")

# Find the minimum value along rows
I_min = np.min(I,1)
print(I_min)
print("-------------------------------------------")

# Find the minimum value along columns
I_min = np.min(I,0)
print(I_min)
print("-------------------------------------------")

# maximum value of the image I
I_max = np.max(I)
print(I_max)
print("-------------------------------------------")

# Find the maximum value along rows
I_max = np.max(I,1)
print(I_max)
print("-------------------------------------------")

# Find the maximum value along columns
I_max = np.max(I,0)
print(I_max)
print("-------------------------------------------")

# Find the mean of the image I
I_mean = np.mean(I)
print("The mean value of I is ", I_mean)
print("-------------------------------------------")

# Calculate mean without using the built-in command
I_mean = np.sum(I)/(M*N)
print("The mean value of I is ",I_mean)
print("-------------------------------------------")
# Calculate the median of the image I
I_median = np.median(I)
print("The median value of I is ", I_median)
```

###  Nearest Neighbor Interpolation

This method is the simplest technique that re samples the pixel values present in the input vector or a matrix. In MATLAB, ‘imresize’ function is used to interpolate the images.

<font color=red>**In the below code, we will implement the nearest neighbor interpolation using our own code.** </font>
***
#### <font color=red> **[OPTIONAL]**</font> Our Implementation 

```python
import numpy as np

I = np.array([[10,4,22],[2, 18, 7],[9, 14, 25]],dtype='uint8')
print(I)
print('-------------------------')
m1,n1 = I.shape
I_new = np.zeros((6,6))
m2, n2 = I_new.shape

Ratio_row, Ratio_column = m2/m1, n2/n1
Row_position = np.floor(np.arange(0,m2)/Ratio_row)
Column_position = np.floor(np.arange(0,n2)/Ratio_column)
xx, yy = np.meshgrid(Row_position,Column_position)
xx, yy = xx.astype(int), yy.astype(int)
xx, yy = xx.T, yy.T
# print(xx)
# print('----------------------')
# print(yy)
for i in range(0,m2):
    for j in range(0,n2):
        I_new[i,j] = I[xx[i,j],yy[i,j]]
        

print(I_new)
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

I = cv2.imread('SampleImage1.jpeg')
I = cv2.cvtColor(I,cv2.COLOR_BGR2GRAY)
m1,n1 = I.shape
print('The size of original image is ',m1,'x',n1)
I_new = np.zeros((2*m1,2*n1))
m2, n2 = I_new.shape
print('The size of the new image should be ',m2,'x',n2)
Ratio_row, Ratio_column = m2/m1, n2/n1
Row_position = np.floor(np.arange(0,m2)/Ratio_row)
Column_position = np.floor(np.arange(0,n2)/Ratio_column)
xx, yy = np.meshgrid(Row_position,Column_position)
xx, yy = xx.astype(int), yy.astype(int)
xx, yy = xx.T, yy.T
for i in range(0,m2):
    for j in range(0,n2):
        I_new[i,j] = I[xx[i,j],yy[i,j]]
        

#plt.subplot(121)
plt.imshow(I,'gray',interpolation='none',aspect='equal')
plt.title('Original Image')
plt.show()
#plt.subplot(122)
plt.imshow(I_new,'gray',interpolation='none',aspect='equal')
plt.title('Resized With Nearest Neighbor Interpolatioin')
plt.show()
I_new.shape
```
